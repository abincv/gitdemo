```python
# 2.How to stack two random series vertically and horizontally?

#vertical stacking
newseries1=pd.Series(np.ones(5))
newseries0=pd.Series(np.zeros(5))
print("vertical stacking\n")
print(pd.concat([newseries1,newseries0], axis = 0))

#horizontal stacking
print("\nhorixontal stacking\n")
print(pd.concat([newseries1,newseries0], axis = 1))
```

```python
#6.How to filter valid emails from a series data using following pattern to be matched?
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'
emails = pd.Series(['books at amazom.com', 'james@egypt.com', 'jones@yahoo.in', 'bigdata@gmail.com'])
print("valid emails:\n\n",emails[emails.str.match(pattern)])
```

```python
#7a
earthquakedf=pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/earthquakes_2014.csv")
#7b
earthquakedf
```

```python
#7 c.Convert time column from object to datetime format [Hint: Use pd.to_datetime()]
earthquakedf.time=pd.to_datetime(earthquakedf.time)
earthquakedf.time
```

```python
#7 d.Rename columns latitude, longitude, depth, mag to  ‘Lat’, ‘Long’, ‘D, M’ column names.
earthquakedf.rename(columns={'latitude':'Lat', 'longitude':'Long', 'depth':'D', 'mag':'M'},inplace=True)
earthquakedf
```

```python
# 7 e.Find out invalid data (the nan values from missing points)
earthquakedf.isna()
```

```python
# 7.f.Drop all nan values from the existing data.
earthquakedf.dropna(how='any')
```

```python
#7 g.Replace the invalid data with central tendencies.
values={}

for column,types in earthquakedf.dtypes.items() :
    if column == 'id':
        values[column]= '1111111'
    elif types == 'object' :
        values[column]=earthquakedf[column].mode()[0]
    else :
        values[column]=earthquakedf[column].mean()

#values

earthquakedf.fillna(value=values)
```
